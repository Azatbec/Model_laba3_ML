```python
missing_data = hotels.isnull().sum()
missing_data = missing_data[missing_data > 0] 
missing_data
```

```python
hotels = hotels.drop(columns=['company'])
hotels
```

```python
top_countries = hotels['country'].value_counts().head(5)

print(top_countries)
```

```python
max_adr_row = hotels.loc[hotels['adr'].idxmax()]
name = max_adr_row['name']
max_adr = max_adr_row['adr']
print(f"Имя человека с наибольшим ADR: {name}, значение ADR: {max_adr}")
```

```python
average_adr = hotels['adr'].mean() 
average_adr_rounded = round(average_adr, 2)
print(f"Среднее значение ADR для всех бронирований: {average_adr_rounded}")
```

```python
hotels['total_nights'] = hotels['stays_in_weekend_nights'] + hotels['stays_in_week_nights']

average_nights = hotels['total_nights'].mean()


average_nights_rounded = round(average_nights, 2)

print(f"Среднее количество ночей для всех бронирований: {average_nights_rounded}")
```

```python
hotels['total_nights'] = hotels['stays_in_weekend_nights'] + hotels['stays_in_week_nights']


hotels['total_booking_amount'] = hotels['adr'] * hotels['total_nights']


average_booking_amount = hotels['total_booking_amount'].mean()

average_booking_amount_rounded = round(average_booking_amount, 2)

print(f"Средняя общая сумма бронирования: {average_booking_amount_rounded}")
```

```python
special_requests = hotels[hotels['total_of_special_requests'] == 5]
result = special_requests[['name', 'email']]
print(result)
```

```python
total_bookings = hotels.shape[0] 
repeated_guest_bookings = hotels[hotels['is_repeated_guest'] == 1].shape[0] 
percentage_repeated_guests = (repeated_guest_bookings / total_bookings) * 100

print(f"Процент бронирований, классифицированных как бронирования повторными гостями: {percentage_repeated_guests:.2f}%")
```

```python
top_5_surnames = hotels['name'].str.split().str[1].value_counts().head(5)
print(top_5_surnames)
```

```python
hotels['total_children_babies'] = hotels['children'] + hotels['babies'] 
max_children_babies = hotels['total_children_babies'].max()
names_with_max_children_babies = hotels[hotels['total_children_babies'] == max_children_babies]['name']
print(names_with_max_children_babies)
```

```python
hotels['area_code'] = hotels['phone-number'].str[:3] 
top_area_codes = hotels['area_code'].value_counts().head(3)
print(top_area_codes)
```

```python
checkins_1_15 = hotels[(hotels['arrival_date_day_of_month'] >= 1) & (hotels['arrival_date_day_of_month'] <= 15)].shape[0]
checkins_1_15
```

```python
# Преобразуем дату заселения в формат datetime
hotels['arrival_date'] = pd.to_datetime(hotels['arrival_date_year'].astype(str) + '-' +
                                         hotels['arrival_date_month'] + '-' +
                                         hotels['arrival_date_day_of_month'].astype(str))

# Создаем новый столбец для дня недели
hotels['day_of_week'] = hotels['arrival_date'].dt.day_name()

# Подсчитываем количество заселений по дням недели
checkins_by_day = hotels['day_of_week'].value_counts().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Преобразуем в DataFrame
checkins_by_day_df = checkins_by_day.reset_index()
checkins_by_day_df.columns = ['Day of Week', 'Number of Check-ins']

print(checkins_by_day_df)
```
